### Estruturas de repetição - While: uma sequência de tarefas executadas equanto uma condição é verdadeira

```python
contador = 0

while contador < 3:
    contador = contador + 1
    if contador == 1:
        print(contador, "item limpo")
    else:
        print(contador, "itens limpos")
        
print('Fim da repetição do bloco while')
```

```python
contador = 0

while True:
    if contador < 10:
        contador = contador + 1
        if contador == 1:
            print(contador, "item limpo")
        else:
            print(contador, "itens limpos")
    else:
        break;
print('Fim da repetição do bloco while')
```

```python
def calcula_media(*args):
    soma = sum(args)
    media = soma / len(args)
    return media
calcula_media(5,6,7)
```

```python
def calcula_media(*args, margem):
    soma = sum(args)
    media = soma / len(args)
    return media + margem
calcula_media(5,6,7, margem=0.3)
```

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return[
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return{
        'title': title,
        'display': display
    }
```

```python
def create_chart(x,y,labels,kind='bar',title=''):
    
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart={
        'type': kind,
        'data':{
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython import display
```
